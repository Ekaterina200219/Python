```python
import numpy as np
from numpy import sin,pi
import matplotlib.pyplot as plt
t = np.linspace(0,2*np.pi,100) #from 0 to 2pi, 100 elements; #Функция linspace() возвращает одномерный массив из указанного количества элементов, значения которых равномерно распределенны внутри заданного интервала.
plt.plot(np.sin(3*t),np.sin(4*t))
plt.axes().set_aspect(1) # отношение масштаба оси ординат к масштабу оси абцисс
plt.show()
```

```python
import numpy as np
from numpy import sin,cos,pi
import matplotlib.pyplot as plt
t = np.linspace(0,8*np.pi,100) #from 0 to 8pi,100 elements
R = 3.5
rsmall = 2
h = 1
plt.plot((R - rsmall)*np.cos(t) + h*np.cos((R - rsmall)/rsmall * t), (R - rsmall)*np.sin(t) - h*np.sin((R - rsmall)/rsmall * t))
plt.axes().set_aspect(1) # y:x
plt.show()
```

```python
import numpy as np
from numpy import sin,cos,pi
import matplotlib.pyplot as plt
t = np.linspace(0,4*np.pi,100) #from 0 to 4pi,100 elements
R = 2
r = 1
plt.plot((R - r)*np.cos(t) + r*np.cos(t - (r/R * t)), (R - r)*np.sin(t) - r*np.sin(t - (r/R * t)))
plt.axes().set_aspect(1) # y:x
plt.show()
```
