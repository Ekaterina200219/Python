* listdir() возвращает список записей для данного каталога.
* Метод os.path.isfile() возвращает True, если данная запись является файлом.
* Оператор yield завершает работу функции, но сохраняя текущее состояние и возвращает только имя записи являющейся файлом.






```
# посмотрим на файлы в корневой директории Colaboratory
n = 0 # счетчик файлов
#path = '/content/'
def files(path):
    for file in os.listdir(path):
        if os.path.isfile(os.path.join(path, file)):
            yield file
for file in files("."):
    n+=1
    print(file)
print("\nКоличество файлов в корневой директории Colaboratory =", n)
```

```
# создадим список из файлов
files_in_Colab = []
for file in files("."):
    files_in_Colab.append(file)
```
