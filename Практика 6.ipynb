```python
# Задание 1.
import numpy as np
def sigmoid(x):
    #Функция активации: f(x) = 1/ (1+e^(-x))
    return 1/ (1 + np.exp(-x))
class Neuron:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias
        
    def feedforward(self, inputs):
        total = np.dot(self.weights, inputs) + self.bias
        return sigmoid(total)
    
weights = np.array([0.5, 0.5, 0.5])
bias = 4
n = Neuron(weights, bias)
x = np.array([2,3,4])
print(n.feedforward(x))
```

```python
class OurNeuralNetwork:
    '''
    Данные нейросети:
    - три входа (x1, x2, x3)
    - три нейрона в скрытых слоях (h1, h2, h3)
    - выход (o1)
    
    Нейроны имеют индентичные веса и пороги:
    - w = [0.5, 0.5, 0.5]
    - b = 0
    '''
    def __init__(self):
        weights = np.array([0.5, 0.5, 0.5])
        bias = 0
        #Класс Neuron из предыдущего раздела
        self.h1 = Neuron(weights, bias)
        self.h2 = Neuron(weights, bias)
        self.h3 = Neuron(weights, bias)
        self.o1 = Neuron(weights, bias)
    def feedforward(self, x):
        out_h1 = self.h1.feedforward(x)
        out_h2 = self.h2.feedforward(x)
        out_h3 = self.h3.feedforward(x)
        # Входы для o1 - это выходы h1, h2, h3
        out_o1 = self.o1.feedforward(np.array([out_h1, out_h2, out_h3]))
        return np.array([out_o1])
        
network = OurNeuralNetwork()
x = np.array([2,3,4])
print(network.feedforward(x))
```

```python
#Задание 5
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import sklearn
```

```python
# MLPClassifier

url = 'https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv'
digits = pd.read_csv(url)
digits
```

```python
data = digits[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
target = digits['variety']
target = target.replace({'Setosa': 0, 'Versicolor': 1, 'Virginica': 2})
```

```python
X_digits, Y_digits = data, target
print('Dataset Sizes: ', X_digits.shape, Y_digits.shape)
```

```python
url1 = 'https://raw.githubusercontent.com/AnnaShestova/salary-years-simple-linear-regression/master/Salary_Data.csv'
salary = pd.read_csv(url1)
salary 
```

```python
data1 = salary[['Salary']]
target1 = salary['YearsExperience']
```

```python
X_salary, Y_salary = data1, target1
print('Dataset Sizes: ', X_salary.shape, Y_salary.shape)
```

```python
# Разделение на обучающую и тестовую выборки
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_digits, Y_digits,
                                                   train_size = 0.80, test_size = 0.20, stratify = Y_digits, random_state = 123)
print('Train/Test Sizes: ', X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
# модель MLPClassifier с параметрами для тренировочных данных
from sklearn.neural_network import MLPClassifier

mlp_classifier = MLPClassifier(random_state=123)
mlp_classifier.fit(X_train, Y_train)
```

```python
Y_preds = mlp_classifier.predict(X_test)

print(Y_preds[:15])
print(Y_test[:15])

## Метод Score для оценки точности моделей классификации.
print('Test Accuracy: %.3f'%mlp_classifier.score(X_test, Y_test))
print('Training Accuracy: %.3f'%mlp_classifier.score(X_train, Y_train))
```

```python
# Матрица путаницы

from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(Y_test, Y_preds):
    conf_mat = confusion_matrix(Y_test, Y_preds)
    fig = plt.figure(figsize = (6,6))
    plt.matshow(conf_mat, cmap = plt.cm.Blues, fignum = 1)
    plt.yticks(range(3), range(3))
    plt.xticks(range(3), range(3))
    plt.colorbar();
    for i in range(3):
        for j in range(3):
                try:
                    plt.text(i-0.2,j+0.1, str(conf_mat[j, i]), color = 'tab:red')
                except IndexError:
                    print("Index out of bounds")
```

```python
plot_confusion_matrix(Y_test, mlp_classifier.predict(X_test))
```

```python
print("Loss: ", mlp_classifier.loss_)
print("Number of Coefs: ", len(mlp_classifier.coefs_))
print("Number of Intercepts: ", mlp_classifier.intercepts_)
print("Number of Iterations for Which Estimator Ran: ", mlp_classifier.n_iter_)
print("Name of Output Layer Activation Function: ", mlp_classifier.out_activation_)
```
